```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import plotly.figure_factory as ff
```

Durante a aula nossa variável a ser prevista (churn) continha apenas 2 categorias, a base de vocês contém mais. O Naive Bayes pode ser aplicado para problemas de classificação com múltiplas classes da mesma forma que para problemas de classificação binária. O Naive Bayes é um algoritmo de classificação probabilístico que calcula a probabilidade de uma amostra pertencer a cada classe e seleciona a classe com a maior probabilidade como a previsão final.
Em resumo, o Naive Bayes pode ser aplicado da mesma maneira para problemas de classificação com múltiplas classes, e os mesmos princípios se aplicam em termos de treinamento, avaliação e aplicação do modelo.

Verifique se o número de linhas condiz, se as variáveis estão corretas sendo apenas a de score para y e as demais nas bases de X e por último, se Y está balanceada no teste.

```python
X_test = pd.read_csv('X_test.csv', delimiter=',')
X_train = pd.read_csv('X_train_balanced.csv', delimiter=',')
y_test = pd.read_csv('y_test.csv', delimiter=',')
y_train = pd.read_csv('y_train_balanced.csv', delimiter=',')
```

```python
train_balance = y_train.value_counts()
print("Balanceamento em y_train:")
print(train_balance)

test_balance = y_test.value_counts()
print("Balanceamento em y_test:")
print(test_balance)
```

Temos os dados balanceados no treino para que não haja viés em nossa análise e desbalanceados no teste, condizendo com os dados reais a serem analisados para avaliação da eficácia do nosso modelo.

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

```python
naive_credit = GaussianNB()
naive_credit.fit(X_train, y_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
y_pred_train = naive_credit.predict(X_train)

accuracy = accuracy_score(y_train, y_pred_train)
print("Acurácia:", accuracy)
```

```python
recall = recall_score(y_train, y_pred_train, average='macro')
print("Recall:", recall)
```

```python
conf_matrix = confusion_matrix(y_train, y_pred_train)

class_names = ['Baixo', 'Médio', 'Alto']

fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='blues',
    showscale=True ,
)

fig.update_layout(
    title='Matriz de confusão',
    xaxis_title='Previsto',
    yaxis_title='Real'
)
fig
```

Uma acurácia de 98% indica que em 98% dos casos o modelo está realizando previsões corretas no conjunto de dados de treinamento, e também o recall em 98% indica que o modelo identifica 98% dos dados foram rotulados de forma correta. Podemos visualizar melhor esses dados na matriz de confusão.

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados, da mesma forma que fez acima. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
y_pred_test = naive_credit.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test, average='macro')
print("Acurácia:", accuracy)
print("Recall:", recall)
```

```python
conf_matrix2 = confusion_matrix(y_test, y_pred_test)

class_names = ['Baixo', 'Médio', 'Alto']

fig2 = ff.create_annotated_heatmap(
    z=conf_matrix2,
    x=class_names,
    y=class_names,
    colorscale='blues',
    showscale=True ,
)

fig2.update_layout(
    title='Matriz de confusão',
    xaxis_title='Previsto',
    yaxis_title='Real'
)
fig2
```

O desempenho do modelo nos dados de teste foi muito próximo do resultado ao aplicarmos na base de treino, o que garante uma boa eficácia. Porém, como os dados foram gerados de forma artificial para a resolução do exercício, é compreensível que as métricas utilizadas para avaliarmos o modelo tenham demonstrado um resultado tão alto, o que dificilmente atingiremos de forma fácil utilizando bases de dados reais ou com tamanhos consideravelmente maiores do que essa.
Podemos considerar que a aplicação e validação do modelo foram feitas de forma positiva uma vez que ele foi capaz de prever com sucesso os rótulos dos dados de teste se baseando nos dados de treino, com resultados semelhantes e sem apresentar classificações erradas.
