Faça a leitura dos dados do arquivo CSV:


Altere o código abaixo de acordo com seu diretório.

```python
df = pd.read_csv("MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv", delimiter=',')

df.head(10)
```

Os campos do nosso dataframe são:

**Title:** Nome do produto.


**Marca:** A marca do produto.


**Preco_Normal:** O preço em que o produto costuma ser vendido quando não há desconto.


**Preco_Desconto:** O preço vendido após o desconto ser aplicado.


**Preco_Anterior:** Preço em que era comercializado o produto antes do desconto aplicado.


**Desconto:** Total de desconto aplicado.






As colunas que aparecem com valores 0 são para os produtos onde não tivemos descontos aplicados.


As categorias estão em espanhol!

# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.
# Identifique as categorias que parecem ter um valor de média abaixo ou acima da mediana.

```python
#Seu código aqui para a média
media = df.groupby('Categoria')['Preco_Normal'].mean()
print(media)
```

```python
#Seu código aqui para a mediana
mediana = df.groupby('Categoria')['Preco_Normal'].median()
print(mediana)
```

Digite aqui as categorias: belleza-y-cuidado-personal,

# 2 - Traga o desvio padrão por categoria de produto.
# Qual o comportamento da média e mediana nas categorias com maior desvio?

```python
#Seu código aqui
desvio = df.groupby('Categoria')['Preco_Normal'].std()
desvio.sort_values(ascending=False)
```

Digite nesse campo o comportamento que você identificou.

A tendência dos valores com maior diferença entre média e mediana é apresentarem um alto desvio padrão, como é o caso da categoria "lacteos". Essa informação nos indica que há uma variação grande entre máximo e mímimo, ou seja, maior dispersão, com presença de outliers.

# 3 - Plot um boxplot da distribuição do Preco_Normal para a categoria que você identificou que tem o maior desvio padrão. Como é a distribuição desses dados segundo o boxplot? Você identifica muitos outliers?

Dica: Para trazer apenas os dados da categoria que você deseja você pode usar o df.loc[df['Categoria'] == 'CATEGORIA ESCOLHIDA'

```python
#Seu código aqui
cat_maior_desvio = desvio.idxmax()

# Filtrando dados
df_cat = df.loc[df['Categoria'] == cat_maior_desvio]

# Plot
plt.figure(figsize=(8,5))
plt.boxplot(df_cat['Preco_Normal'], vert=False)
plt.title(f'Distribuição de Preço Normal - {cat_maior_desvio}')
plt.xlabel('Preço Normal')
plt.show()
```

Como podemos ver na dispersão do gráfico, a maior parte dos valores dessa categoria se encontram com preço abaixo de 5000, com diversos outliers no intervalo entre 5000 e 20000. Essa dispersão justifica o alto valor de desvio padrão.

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui
media_desc = df.groupby('Categoria')['Desconto'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
media_desc.plot(kind='bar', color='skyblue')
plt.title('Média de Descontos por Categoria')
plt.ylabel('Desconto médio')
plt.xlabel('Categoria')
plt.show()
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui
grouped_df = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()

fig = px.scatter(
    grouped_df,
    x='Categoria',
    y='Marca',
    size='Desconto',
    color='Categoria',
    hover_data=['Desconto']
)
fig.update_layout(title='Média de Descontos por Categoria e Marca')
fig.show()
```
